```python
import pandas as pd

df = pd.read_csv('.output.hs300_yearly_financial_report.csv').fillna(0)
df
```

```python
cols = [item for item in df.columns if item not in ['market', 'code', 'year']]
cols
```

```python
df[cols]
```

```python
df_corr = df[cols].corr()
df_corr.stack().reset_index(name="correlation")           
```

```python
import seaborn as sns

g = sns.relplot(
    data=df_corr.stack().reset_index(name="correlation")           ,
    x="level_0", y="level_1", hue="correlation", size="correlation",
    palette="vlag", hue_norm=(-1, 1), edgecolor=".7",
    height=10, sizes=(50, 250), size_norm=(-.2, .8),
)

# Tweak the figure to finalize
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
for artist in g.legend.legendHandles:
    artist.set_edgecolor(".7")
```
